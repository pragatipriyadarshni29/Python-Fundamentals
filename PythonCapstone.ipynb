## 📦 Import Required Libraries 

```python
# Import pandas for data manipulation (DataFrames, CSV, etc.)
import pandas as pd

# Import numpy for numerical operations
import numpy as np
```

---
## ✅ Task 1: Create Three DataFrames and Save as CSV Files

```python
# ── EMPLOYEE DATAFRAME ──────────────────────────────────────────────────────

# Create a dictionary with employee data as given in the project document
employee_data= {
    'ID':     ['A001', 'A002', 'A003', 'A004', 'A005'],                  # Unique employee IDs
    'Name':   ['John Alter', 'Alice Luxumberg', 'Tom Sabestine',         # Full names
               'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],                                 # M = Male, F = Female
    'City':   ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],        # Project locations
    'Age':    [25, 27, 29, 31, 30]                                       # Years the project is active
}

# Convert the dictionary into a pandas DataFrame
employee_df = pd.DataFrame(employee_data)

# Save the Employee DataFrame to a CSV file (index=False avoids writing row numbers)
employee_df.to_csv('employee.csv', index=False)

# Display the Employee DataFrame to verify
print("Employee DataFrame:")
print(employee_df)
print()
```

```python
# ── SENIORITY LEVEL DATAFRAME ────────────────────────────────────────────────

# Create a dictionary with designation (seniority) levels per employee
# Scale: 1 = highest, 4 = lowest; above 4 = loses eligibility

seniority_data= {
    'ID': ['A001','A002','A003','A004','A005'],                     #Employee id
    'Designation Level': [2,2,3,2,3]                                #Current designation levels
}

# Convert the dictionary into a pandas DataFrame
seniority_df = pd.DataFrame(seniority_data)

# Save the Seniority Level DataFrame to a CSV file
seniority_df.to_csv('seniority.csv', index=False)

# Display the Seniority DataFrame to verify
print("Seniority Level DataFrame:")
print(seniority_df)
print()
```

```python
# ── PROJECT DATAFRAME ────────────────────────────────────────────────────────

# Create a dictionary with project data
project_data = {
    'ID':      ['A001','A002','A003','A004','A005',                # Employee IDs for each project
                'A002','A005','A003','A001','A003',
                'A001','A004','A004','A005'],
    'Project': ['Project 1','Project 2','Project 3',               # Project names
                'Project 4','Project 5','Project 6',
                'Project 7','Project 8','Project 9',
                'Project 10','Project 11','Project 12',
                'Project 13','Project 14'],
    'Cost':    [1002000, 2000000, 4500000, 5500000, np.nan,        # np.nan = missing value
                680000, 400000, 350000, np.nan, 300000,
                2000000, 1000000, 3000000, 200000],
    'Status':  ['Finished','Ongoing','Finished','Ongoing',         # Project statuses
                'Finished','Failed','Finished','Failed',
                'Ongoing','Finished','Failed','Ongoing',
                'Finished','Finished']
}
# Note: Some 'Cost' values are missing (NaN = Not a Number) — will be handled in Task 2

# Convert the dictionary into a pandas DataFrame
project_df = pd.DataFrame(project_data)

# Convert the Project DataFrame to CSV file
project_df.to_csv('project.csv', index=False)

# Display the Project DataFrame to verify
print("Project DataFrame:")
print(project_df)
print()

print("✅ Task 1 Complete — All 3 CSV files saved successfully!")
```

## ✅ Task 2 — Fill Missing Cost Values Using Running Average (For Loop)

```python
# Load the project CSV file
project_df = pd.read_csv('project.csv')

# Start both counters at zero
running_total = 0                      #a box that keeps adding up all the costs it has seen so far
running_count = 0                      #counts how many valid costs have been seen so far

# Loop through each row one by one
for i in range(len(project_df)):

    # Check if the cost value in this row is missing
    if pd.isna(project_df.loc[i, 'Cost']):

        # Fill it with the running average calculated so far
        project_df.loc[i, 'Cost'] = running_total / running_count

    else:
        # Value is not missing — add it to our running total
        running_total += project_df.loc[i, 'Cost']
        running_count += 1

# Convert Cost column to integer to remove decimals
project_df['Cost'] = project_df['Cost'].astype(int)

# Print to verify
print(project_df[['Project', 'Cost']])

# Save back to CSV
project_df.to_csv('project.csv', index=False)

print()
print("✅ Task 2 Complete — Missing costs filled using running average!")
```

## ✅ Task 3 — Split 'Name' into 'First Name' and 'Last Name'

```python
#Load the Employee CSV from Task 1
employee_df = pd.read_csv('employee.csv')

#Display original data before splitting
print("Before splitting Name column:")
print(employee_df)
print()

# Split the 'Name' column by a space character into two parts
name_split = employee_df['Name'].str.split(' ',expand=True)              #Expand=True: returns a DataFrame with 2 columns: [0] and [1]

#Assign first part (index=0) as 'FirstName' and second part (index=1) as 'lastName'
employee_df['FirstName']= name_split[0]
employee_df['LastName']= name_split[1]

# Remove the original 'Name' column since we no longer need it
employee_df = employee_df.drop(columns=['Name'])

#Display the updated DataFrame with the new split columns 
print('After splitting name column:')
print(employee_df)
print()

# Save the updated employee DataFrame back to CSV
employee_df.to_csv('employee.csv', index=False)

print()
print("✅ Task 3 Complete — Name column split into First Name and Last Name!")
```

---
## ✅ Task 4 — Join All Three DataFrames into One ('Final')

```python
#Load all three updated CSV files
employee_df = pd.read_csv('employee.csv')              #contains employee personal details
seniority_df = pd.read_csv('seniority.csv')            #contains designation levels
project_df = pd.read_csv('project.csv')                #contains project details 

#Step1: Merge employee_df and seniority_df on 'ID' column using INNER JOIN to keep only matching IDs from both tables
merged_emp_sen = pd.merge(employee_df, seniority_df, on='ID', how='inner')

#Step2: Merge the result with project_df on 'ID' column using INNER JOIN again
Final = pd.merge(merged_emp_sen, project_df, on='ID', how='inner')

#Display the final merged DF
print('Final DataFrame (all 3 tables joined):')
print(Final)
print()

print("✅ Task 4 Complete — All three DataFrames merged into 'Final'!")
```

---
## ✅ Task 5 — Add Bonus Column (5% of Cost for 'Finished' Projects)

```python
#Add a new 'Bonus' Column to the Final DF
#Rule: 5% of 'Cost' only if Status == 'Finished' ; otherwise, bonus is 0

#np.where(condition, value_if_true, value_if_false) should be applied 
Final['Bonus'] = np.where(
    Final['Status'] == 'Finished',         #Condition: Project status is 'Finished'
    Final['Cost'] * 0.05,                  # If True: bonus= 5% of Cost
    0                                      #If False: bonus= 0
)

#Display the Final DF with new Bonus Column
print('Final DF with Bonus column')
print(Final[['ID','FirstName','Project','Cost','Status','Bonus']])

print()
print("✅ Task 5 Complete — Bonus column added (5% for Finished projects)!")
```

---
## ✅ Task 6 — Demote Designation for Failed Projects; Remove if Level > 4

```python
#Display designation levels before demotion
print('Before demotion:')
print(Final[['ID','FirstName','Project','Status','Designation Level']])
print()

#Step1: Demote(+1) for rows where Status is 'Failed'
# Higher number = lower rank i.e., (1 is highest, 4 is lowest)
Final.loc[Final['Status']== 'Failed','Designation Level'] += 1

#Step2: Delete rows where Designation Level > 4 (these employees lose eligibility)
Final = Final[Final['Designation Level'] <= 4]    # Keep only rows where level is 4 or below

#Reset index after deleting rows so it stays clean and sequential
Final= Final.reset_index(drop=True)

#Display the cleaned DF
print('After removing employees with Designation Level > 4:')
print(Final[['ID','FirstName','Project','Status','Designation Level']])

print()
print("✅ Task 6 Complete — Failed project employees demoted; ineligible records removed!")
```

---
## ✅ Task 7 — Add 'Mr.' / 'Mrs.' to First Name and Drop Gender Column

```python
#Dispay before adding title
print('Before adding title prefix:')
print(Final[['FirstName','Gender']])
print()

#Add 'Mr.' prefix for males (Gender==M) and 'Mrs.' prefix for females (Gender==F)
#Using np.where to apply the condition row by row
Final['FirstName']= np.where(
    Final['Gender']== 'M',                           # Condition checks if Gender is male
    'Mr. '+ Final['FirstName'],                      # If True: add Mr. prefix
    'Mrs. ' + Final['FirstName']                     # If False (i.e., if Female): add Mrs. prefix
)

#Drop the 'gender' column since it's no longer required as title already reflects it
Final= Final.drop(columns=['Gender'])

#Display updated FirstName column
print ('After adding title prefix and delteing Gender column:')
print(Final[['ID','FirstName','LastName']])

print()
print("✅ Task 7 Complete — Titles added and Gender column dropped!")
```

---
## ✅ Task 8 — Promote Designation Level by 1 for Employees Aged > 29

```python
# Display before promotion
print("Before promotion (Age > 29):")
print(Final[['ID', 'FirstName', 'Age', 'Designation Level']].drop_duplicates())
print()

# Using IF condition to promote designation level
for i in Final.index:                            # Loop through every row index in Final
    if Final.loc[i, 'Age'] > 29:                 # IF the employee's age is more than 29
        Final.loc[i, 'Designation Level'] -= 1   # Decrease level by 1 (promote)

# Display after promotion
print("After promotion (employees with Age > 29 get -1 to Designation Level):")
print(Final[['ID', 'FirstName', 'Age', 'Designation Level']].drop_duplicates())

print()
print("✅ Task 8 Complete — Designation level promoted for employees aged > 29!")
```

---
## ✅ Task 9 — Create 'TotalProjCost' DataFrame (Total Cost per Employee)

```python
#Group the Final DF by 'ID', 'FirstName' and sum the 'Cost' for each emplyee
#groupby() groups all rows with the same ID together; sum() addds up their costs
TotalProjCost = Final.groupby(['ID','FirstName'], as_index=False) ['Cost'].sum()

#Rename the 'Cost' column to 'Total Cost' as required
TotalProjCost= TotalProjCost.rename(columns={'Cost': 'Total Cost'})

#Keeping only the required 3 columns: ID, FirstName, Total cost
TotalProjCost= TotalProjCost[['ID','FirstName','Total Cost']]

#Display the TotalProjCost DF
print('TotalProjCost DF:')
print(TotalProjCost)

print()
print("✅ Task 9 Complete — TotalProjCost DataFrame created with total cost per employee!")
```

## ✅ Task 10 — Print Employee Details Where City Contains the Letter 'o'

```python
#Filter rows from the Final DF where the 'City' column contains the letter 'o'
city_with_o = Final[Final['City'].str.contains('o', case=False)]

#Dropping duplicate rows to avoid showing same employee multiple times due to multiple projects
city_with_o_unique= city_with_o.drop_duplicates(subset=['ID'])                 

#Displayy employees whose city name contains the letter 'o'
print("Employee Details Where City Contains the Letter 'o':")
print(city_with_o_unique[['ID','FirstName','LastName','City','Age','Designation Level']])

print()
print("✅ Task 10 Complete — Employees from cities containing 'o' displayed!")
```
